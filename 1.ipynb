```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def spatial_averaging_filter(image, kernel_size):
    kernel = np.ones((kernel_size, kernel_size), dtype=np.float32) / (kernel_size ** 2)
    return cv2.filter2D(image, -1, kernel)

def main():
    # Load the image using cv2
    image_path = 'F:/Nitheesh/Computer vision/Exercise4/image1.png'  # Replace with your image file name
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    # Apply spatial averaging filter with kernel size 3x3
    kernel_size = 3
    filtered_image = spatial_averaging_filter(image, kernel_size)

    # Display the original and filtered images
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(image, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(filtered_image, cmap='gray')
    plt.title('Filtered Image')
    plt.axis('off')

    plt.show()

```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

def spatial_average_filter(image, kernel_size):
    height, width = image.shape
    padded_image = np.pad(image, pad_width=kernel_size//2, mode='constant', constant_values=0)
    result_image = np.zeros_like(image)

    for y in range(height):
        for x in range(width):
            sum_val = 0
            for i in range(kernel_size):
                for j in range(kernel_size):
                    sum_val += padded_image[y + i, x + j]
            result_image[y, x] = sum_val // (kernel_size ** 2)

    return result_image
image = cv2.imread('F:/Nitheesh/Computer vision/Exercise4/image1.png', cv2.IMREAD_GRAYSCALE)
kernel_size = 5
blurred_image = spatial_average_filter(image, kernel_size)

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(blurred_image, cmap='gray')
plt.title('Blurred Image')
plt.axis('off')

plt.show()
```
