```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

def plot_histogram(image, title):
    histogram = cv2.calcHist([image], [0], None, [256], [0, 256])

    histogram_normalized = histogram / np.sum(histogram)

    cdf = histogram.cumsum()
    cdf_normalized = cdf / cdf[-1]

    height, width = image.shape

    plt.figure(figsize=(10, 6))

    plt.subplot(2, 2, 1)
    plt.imshow(image, cmap='gray')
    plt.title(title)
    plt.axis('off')

    plt.subplot(2, 2, 2)
    plt.plot(histogram)
    plt.title('Original Histogram')
    plt.xlabel('Pixel Intensity')
    plt.ylabel('Frequency')

    plt.subplot(2, 2, 3)
    plt.plot(histogram_normalized)
    plt.title('Normalized Histogram')
    plt.xlabel('Pixel Intensity')
    plt.ylabel('Normalized Frequency')

    plt.subplot(2, 2, 4)
    plt.plot(cdf_normalized * 255)
    plt.title('Equalized Histogram')
    plt.xlabel('Pixel Intensity')
    plt.ylabel('Equalized Intensity')

    plt.tight_layout()
    plt.show()

image = cv2.imread('F:/Nitheesh/Computer vision/Exercise4/image1.png', cv2.IMREAD_GRAYSCALE)

plot_histogram(image, 'Original Image')

image = cv2.imread('F:/Nitheesh/Computer vision/Exercise4/image2.png', cv2.IMREAD_GRAYSCALE)

plot_histogram(image, 'Original Image')
```
