```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
def spatial_median_filter(image, kernel_size):
    height, width = image.shape
    padded_image = np.pad(image, pad_width=kernel_size//2, mode='constant', constant_values=0)
    result_image = np.zeros_like(image)

    for y in range(height):
        for x in range(width):
            neighbors = []
            for i in range(kernel_size):
                for j in range(kernel_size):
                    neighbors.append(padded_image[y + i, x + j])
            neighbors.sort()
            result_image[y, x] = neighbors[kernel_size**2 // 2]

    return result_image

image = cv2.imread('F:/Nitheesh/Computer vision/Exercise4/image1.png', cv2.IMREAD_GRAYSCALE)

kernel_size = 5

filtered_image = spatial_median_filter(image, kernel_size)

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(filtered_image, cmap='gray')
plt.title('Filtered Image')
plt.axis('off')

plt.show()
```
